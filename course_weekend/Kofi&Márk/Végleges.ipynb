## 8) melyik az a cset, ahol a legtöbb idő telt el két üzenet között, és mennyi ez az idő?

```python
#Létrehozok egy olyan könyvtárat (egyedi), amelyben az szerepel, 
#hogy az adott chatbe milyen időpontokban érkeztek be üzenetek.

egyedi ={}
for i in range(2010,2021):
    for dic in mega[i]:
        if dic["type"] == "Generic":
            if dic["thread_path"]in egyedi.keys():
                egyedi[dic["thread_path"]].append(dic["timestamp_ms"])
            else:
                egyedi[dic["thread_path"]]=[dic["timestamp_ms"]]
```

```python
#Létrehozok egy olyan könyvtárat (különbség), amiben az szerepel, 
#hogy adott chatbe milyen időeltérésekkel érkeztek be az üzenetek
#Aztán pedig az adott chat időeltérés maximumát hagyom csak meg különbségek nevű könyvtárban.

különbségek={}
for key in egyedi.keys():
    timestamplista=egyedi[key]
    timestamplista.sort()
    timestamplista.reverse()
    különbség=[timestamplista[i]-timestamplista[i+1] for i in range(len(timestamplista)-1)]
    if len(különbség)==0:
        pass
    else:
        legnagyobb=max(list(különbség))
    különbségek[key]=legnagyobb

{k: v for k, v in sorted(különbségek.items(), key=lambda item: item[1], reverse=True)}
```
