## Hányan vannak, akik minden évben küldtek üzenetet? (3 pont)

````python
nevek = {}

for y in range(2010,2020):
    list_of_dicts = get_data_by_year(y)
    for d in list_of_dicts:    
        if d['sender_name'] in nevek:
            if y not in nevek[d['sender_name']]:
                nevek[d['sender_name']].append(y)         
        else:
            nevek[d['sender_name']] = [y]
            
def leghosszabb(elem):
    return len(elem[1])
nevek = sorted(nevek.items(), key = leghosszabb, reverse= True)
print(nevek)
````
